```python
import numpy as np
import math
from matplotlib import pyplot as plt
import numpy.linalg as la
```

# Problem 1

## a)

```python
def genV(xint):
    n = len(xint)
    v = np.empty((n,n))
    
    for i in range(n):
        for j in range(n):
            v[j][i] = xint[j]**(n-i-1)
    
    return v

def evalpoly(xeval, coefs, N):
    
    '''evaluate polynomial'''
    yeval = 0.
    
    for j in range(N):
        yeval += coefs[j]*(xeval**(N-j-1))
    
    return yeval
```

## b)

```python
def driver(N):
    
    f = lambda x: 1/(1+(10*x)**2)
    
    ''' interval'''
    a = -1
    b = 1
    
    ''' create interpolation nodes'''
    xint = np.empty(N)
    
    for j in range(1, N+1):
        xint[j-1] = -1 + (j-1)*(2/(N-1))
    
    ''' create interpolation data'''
    yint = f(xint)
    
    ''' create points for evaluating the interpolating polynomial'''
    Neval = 1000
    xeval = np.linspace(a,b,Neval+1)
    yeval= np.zeros(Neval+1)
    
    ''' gen coefs '''
    v = genV(xint)
    vinv = la.inv(v)
    coefs = vinv.dot(yint)
    
    ''' evaluate poly '''
    for kk in range(Neval+1):
        yeval[kk] = evalpoly(xeval[kk],coefs,N)
    
    
    ''' create vector with exact values'''
    fex = f(xeval)
    
    return xeval, fex, yeval, xint, yint
```

```python
col = ['b','g','r','c','m','y']
i = 0

for n in range(2,20,4):
    fig = plt.figure(figsize=(10,6))
    ax = plt.subplot(121)
    xeval, fex, yeval, xint, yint = driver(n)
    plt.plot(xeval,fex,'k-', label='Exact Values')
    plt.plot(xeval,yeval, col[i], label='P(x), N = '+str(n))
    plt.plot(xint,yint,'o'+col[i], label='Interp, N = '+str(n))
    ax.legend()
    
    ax = plt.subplot(122)
    err = abs(yeval-fex)
    plt.semilogy(xeval,err,col[i],label='Absolute Error, N = '+str(n))
    ax.legend()
    
    i +=1
    
    plt.show()
```

#### We see that the error at the ends of the interval is much higher than at the middle due to Runge's phenomenon.

# Problem 2

```python
def barycentric(xstar, xint, yint, N):
   
    phi = 1
    w = np.ones(N)
    
    for j in range(N):
        phi = phi*(xstar-xint[j])
        
        for i in range(N):
            if not i == j:
                w[j] = w[j]*(xint[j] - xint[i])
                
        w[j] = 1/w[j]
    
    s = 0
    for k in range(N):
        s += (w[k]*yint[k])/(xstar - xint[k])
    
    return phi*s
```
