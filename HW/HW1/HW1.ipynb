```python
x = np.arange(1.920, 2.081, 0.001)
p1 = x**9 - 18*(x**8) + 144*(x**7) - 672*(x**6) + 2016*(x**5) - 4032*(x**4) + 5376*(x**3) - 4608*(x**2) + 2304*x - 512
p2 = (x-2)**9

plt.plot(x, p1)
plt.plot(x, p2)

plt.legend(["(i) Expanded poly", "(ii) (x-2)^9"])
plt.show()
```
